```python
import requests
import json
import sseclient

BASE_URL = "http://127.0.0.1:8000"
APP_NAME = "travel_advisor"
USER_ID = "u_123"

```

```python
response = requests.post(f"{BASE_URL}/apps/{APP_NAME}/users/{USER_ID}/sessions")

print(response.json())

SESSION_ID = response.json()["id"] or ""
```

```python

message = {
    "appName" : APP_NAME,
    "userId" : USER_ID,
    "sessionId" : SESSION_ID,
    "newMessage" : {
        "parts" : [
            {
                "text" : "안녕 나는 파리랑 서울에 놀러 가고 싶어 두 도시에 대해서 비교해줘"
            }
        ],
        "role" : "user"
    },
    "streaming" : True,
}

response = requests.post(f"{BASE_URL}/run_sse", json=message,stream=True,)

client = sseclient.SSEClient(response)

for event in client.events():
    data = json.loads(event.data)

    content = data.get("content") or {}
    parts = content.get("parts") or []

    for part in parts:
        function_call = part.get("functionCall", None)

        if function_call:
            print(function_call.get("name", "unknown"))
        
        text = part.get("text", None)

        if text:
            print(text)
    print("=" * 60)
```
