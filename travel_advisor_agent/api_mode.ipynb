```python
import requests

BASE_URL = "http://127.0.0.1:8000"
APP_NAME = "travel_advisor"
USER_ID = "u_123"

```

```python
response = requests.post(f"{BASE_URL}/apps/{APP_NAME}/users/{USER_ID}/sessions")

print(response.json())

SESSION_ID = response.json()["id"] or ""
```

```python
message = {
    "appName" : APP_NAME,
    "userId" : USER_ID,
    "sessionId" : SESSION_ID,
    "newMessage" : {
        "parts" : [
            {
                "text" : "30달러"
            }
        ],
        "role" : "user"
    }
}

response = requests.post(f"{BASE_URL}/run", json=message)

data = response.json()

for event in data:
    content = event["content"]
    parts = content["parts"]

    for part in parts:
        function_call = part.get("functionCall", None)

        if function_call:
            print(function_call.get("name", "unknown"))
        
        text = part.get("text", None)

        if text:
            print(text)
    print("=" * 60)
```
